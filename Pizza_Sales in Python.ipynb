```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\Tosin\Downloads\pizza_sales.csv")
df
```

```python
#Based on the format in the above data, the date format is 'DD-MM-YYYY'. 
#To Update let use "pd.read_csv" line to include the parse_dates parameter with the appropriate format:
# Assuming 'order_date' is the column containing date information
df = pd.read_csv(r"C:\Users\Tosin\Downloads\pizza_sales.csv", parse_dates=['order_date'], dayfirst=True)
df
```

```python
# 1. Total Revenue
total_revenue = df['total_price'].sum()

# 2. Average Order Value
average_order_value = df.groupby('order_id')['total_price'].sum().mean()

# 3. Total Pizza Sold
total_pizza_sold = df['quantity'].sum()

# 4. Total Orders
total_orders = df['order_id'].nunique()

# 5. Average Pizza per Order (rounded to 2 decimal places)
average_pizza_per_order = df.groupby('order_id')['quantity'].sum().mean()

# Display the results
print("1. Total Revenue:", total_revenue)
print("2. Average Order Value:", average_order_value)
print("3. Total Pizza Sold:", total_pizza_sold)
print("4. Total Orders:", total_orders)
print("5. Average Pizza per Order (rounded to 2 decimal places): {:.2f}".format(average_pizza_per_order))
```

```python
# Convert 'order_date' to datetime type
df['order_date'] = pd.to_datetime(df['order_date'])

# 6. Daily Trend for Total Orders
daily_trend = df.groupby(df['order_date'].dt.strftime('%A'))['order_id'].nunique()

# 7. Monthly Trend for Total Orders
monthly_trend = df.groupby(df['order_date'].dt.strftime('%B'))['order_id'].nunique()

# Display the results
print("\n6. Daily Trend for Total Orders:\n", daily_trend)
print("\n7. Monthly Trend for Total Orders:\n", monthly_trend)
```

```python

# 8. Percentage of Sales by Pizza Category
sales_by_category_percentage = (df.groupby('pizza_category')['total_price'].sum() / total_revenue * 100).round(2)

# 9. Percentage of Sales by Pizza Size
sales_by_size_percentage = (df.groupby('pizza_size')['total_price'].sum() / total_revenue * 100).round(2)

# 10. Total Pizza Sold by Pizza Category
total_pizza_sold_by_category = df.groupby('pizza_category')['quantity'].sum()

# 11. Top 5 Best Sellers by Revenue, Quantity, and Total Orders respectively
top_sellers_revenue = df.groupby('pizza_name')['total_price'].sum().nlargest(5)
top_sellers_quantity = df.groupby('pizza_name')['quantity'].sum().nlargest(5)
top_sellers_orders = df.groupby('pizza_name')['order_id'].nunique().nlargest(5)

# 12. Bottom 5 Sellers by Revenue, Quantity, and Total Orders respectively
bottom_sellers_revenue = df.groupby('pizza_name')['total_price'].sum().nsmallest(5)
bottom_sellers_quantity = df.groupby('pizza_name')['quantity'].sum().nsmallest(5)
bottom_sellers_orders = df.groupby('pizza_name')['order_id'].nunique().nsmallest(5)

# Display the results

print("\n8. Percentage of Sales by Pizza Category:\n", sales_by_category_percentage)
print("\n9. Percentage of Sales by Pizza Size:\n", sales_by_size_percentage)
print("\n10. Total Pizza Sold by Pizza Category:\n", total_pizza_sold_by_category)
print("\n11. Top 5 Best Sellers by Revenue:\n", top_sellers_revenue)
print("\nTop 5 Best Sellers by Quantity:\n", top_sellers_quantity)
print("\nTop 5 Best Sellers by Total Orders:\n", top_sellers_orders)
print("\n12. Bottom 5 Sellers by Revenue:\n", bottom_sellers_revenue)
print("\nBottom 5 Sellers by Quantity:\n", bottom_sellers_quantity)
print("\nBottom 5 Sellers by Total Orders:\n", bottom_sellers_orders)
```
